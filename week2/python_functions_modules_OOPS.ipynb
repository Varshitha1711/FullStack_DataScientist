```python
# Python functions
# 1. Fibonacci Generator
# * Function generates n Fibonacci numbers.
# * Input: n = 7
# * Output: [0, 1, 1, 2, 3, 5, 8]

def fibonacci(n):
    f0=0
    f1=1
    for _ in range(n):
        print(f0,end=' ')
        f0,f1=f1,f0+f1

n=int(input())
fibonacci(n)
```

```python
# 2. Factorial Using Recursion
 
# * Compute factorial using recursion.
 
# * Input: 5
 
# * Output: 120
 
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
n=int(input())
print(factorial(n))
```

```python
# 3. Unique Words Counter
 
# * Return unique words in a sentence.
 
# * Input: "apple banana apple orange banana"
 
# * Output: {'apple':2,'banana':2,'orange':1}

def word_count(items):
    l=items.split()
    words={}
    for i in l:
        if i in words.keys():
            words[i]+=1
        else:
            words[i]=1
    return words
words=input()
print(word_count(words))
```

```python
# 4. Grade Calculator
 
# * Function takes marks and returns grade.
 
# * Input: [85, 90, 78]
 
# * Output: "Grade: A"
 
def grade(marks):
    s=sum(marks)
    av=s/len(marks)
    if av>75:
        return "A"
    elif av>60:
        return "B"
    elif av>40:
        return "P"
    else:
        return "F"
marks=list(map(int,input().split()))
print("Grade : ",grade(marks))
```

```python
# 5. Second Largest Number
 
# * Return 2nd largest number from a list.
 
# * Input: [10, 5, 20, 8] 
# * Output: 10

def second_max(nums):
    first=sec=float("-inf")
    for num in nums:
        if num>first:
            sec=first
            first=num
        elif num<first and num>sec:
            sec=num
    return sec

nums=list(map(int,input().split()))
print(second_max(nums))
```

```python
# ## Python Modules ( 5 challenges)
 
# 1. os → Count Files by Extension
 
# Write a program using os to count how many .txt and .py files are in the current directory.
 
# Sample Input (files in folder):
# ["notes.txt", "app.py", "data.csv", "report.txt", "main.py"]

# Sample Output:
 
# Text files: 2
 
# Python files: 2

import os
def count_files():
    text=0
    python=0
    for file in os.listdir("."):
        if file.endswith("txt"):
            text+=1
        elif file.endswith("py"):
            python+=1
    print(f"Text files : {text}")
    print(f"Python file :{python}")
count_files()
```

```python
# 2. statistics → Exam Score Analysis
 
# Use statistics to calculate the mean, median, and variance of student exam scores.
 
# Sample Input:
 # Scores = [75, 80, 90, 100, 85, 90, 95]

# Sample Output:
 
# Mean = 87.86
# Median = 90
# Variance = 73.81
 
import statistics
def calculate(scores):
    mea=statistics.mean(scores)
    med=statistics.median(scores)
    var=statistics.variance(scores)
    print(f"Mean = {mea:.2f}")
    print(f"Median = {med}")
    print(f"Variance = {var:.2f}")
scores = [75, 80, 90, 100, 85, 90, 95]
calculate(scores)
```

```python
# 3. math → Quadratic Equation Solver
 
# Use the math module to solve quadratic equations of the form ax² + bx + c = 0.
 
# Sample Input:
 
# a = 1, b = -3, c = 2
 
# Sample Output: 
# Roots are: 1.0 and 2.0
import math
def quadratic(a,b,c):
    d=b*b-4*a*c
    if d > 0:
        root1 = (-b + math.sqrt(d)) / (2*a)
        root2 = (-b - math.sqrt(d)) / (2*a)
        print(f"Roots are: {root1} and {root2}")
    elif d == 0:
        root = -b / (2*a)
        print(f"Root is: {root}")
    else:
        real = -b / (2*a)
        imag = math.sqrt(-d) / (2*a)
        print(f"Roots are: {real}+{imag}i and {real}-{imag}i")
quadratic(1,-3,2)
```

```python
# 4. random → Rock Paper Scissors Game
 
# Use random.choice() to let the computer play Rock–Paper–Scissors against the user.
 
# Sample Input:

# User: Rock
 
# Computer (random): Scissors

# Sample Output:
 
# You chose Rock, Computer chose Scissors
 
# You Win!

import random
def rock_paper_scissors():
    options=["Scissors","Paper","Rock"]
    user=input()
    if user not in options:
        print("Invalid choice")
        return
    com=random.choice(options)
    print(f"You chose {user}, computer chose {com}")
    if user==com:
        print("It's a tie")
    elif(user=='Scissors' and com=='Paper')or (user=='Rock' and com=='Scissors') or (user=='Paper' and com=='Rock'):
        print("You Win!")
    else:
        print("You lose!")
rock_paper_scissors()
```

```python
# Python OOPs
 
# ### 1. Library System
 
# Create a Library class with:
 
# *Attributes: books (dict with book title → copies available)
 
# *Methods:
 
# * borrow(title) → decreases stock if available 
# * return_book(title) → increases stock
# * show_books() → prints all available books
 
# Sample Input:
# python
 
# lib = Library({"Python 101": 3, "Data Science": 2})
# print(lib.borrow("Python 101"))
# print(lib.return_book("Python 101"))
# lib.show_books()

# Sample Output:
 
# You borrowed Python 101
# You returned Python 101
# Available books: {'Python 101': 3, 'Data Science': 2}


class Library:
    def __init__(self,items):
        self.books=items
    def borrow(self,title):
        for book,val in self.books.items():
            if book==title and val>0:
                self.books[book]-=1
                print(f"You borrowed {title}")
                return
        else:
            print("Book not avaliable")
    def return_book(self,title):
        for book,val in self.books.items():
            if book==title:
                self.books[book]+=1
            else:
                self.books[book]=1
        print(f"You returned {title}")
    def show_books(self):
        print("Available books are:")
        for book,val in self.books.items():
            print(f"{book} : {val}")

lib = Library({"Python 101": 3, "Data Science": 2})
lib.borrow("Python 101")
lib.return_book("Python 101")
lib.show_books()
```

```python
# ### 2. ATM Machine with PIN
 
# Create an ATM class with:
 
# *Attributes: balance, pin, is_authenticated
 
# *Methods:
 
# * login(pin) → checks PIN
 
# * check_balance()
 
# * deposit(amount)
 
# * withdraw(amount)
 
# Sample Input:
 
# python
 
# atm = ATM(1234, 500)
 
# print(atm.login(1234))
 
# print(atm.deposit(200))
 
# print(atm.withdraw(100))
 
# print(atm.check_balance())
# Sample Output:
 

 
# Access Granted
 
# Deposited 200
 
# Withdrew 100
 
# Balance: 600


class ATM:
    def __init__(self,pin,balance):
        self.balance=balance
        self.pin=pin
    def login(self,pin):
        if (self.pin==pin):
            return ("Access Granted")
        else:
            return ("Access Denied")
    def deposit(self,amt):
        self.balance+=amt
        return (f"Deposited {amt}")
    def withdraw(self,amt):
        self.balance-=amt
        return (f"Withdrew {amt}")
    def check_balance(self):
        return (f"Balance: {self.balance}")

atm = ATM(1234, 500)
 
print(atm.login(1234))
 
print(atm.deposit(200))
 
print(atm.withdraw(100))
 
print(atm.check_balance())   
```

```python
# 3. University System
 
# Create Professor and Student classes.
 
# *Professor Attributes: name, subject
 
# *Professor Methods: teach(course), give_assignment(task)
 
# *Student Attributes: name, courses, assignments
 
# *Student Methods: enroll(course), submit_assignment(task)
 
# Sample Input:
 
# python
 
# prof = Professor("Dr. Smith", "Computer Science")
# stud = Student("Alice")
# print(prof.teach("Python"))
# print(stud.enroll("Python"))
# print(prof.give_assignment("Project 1")) 
# print(stud.submit_assignment("Project 1"))
 
# Sample Output:

# Dr. Smith is teaching Python
 
# Alice enrolled in Python
 
# Dr. Smith assigned: Project 1
 
# Alice submitted: Project 1

class Professor:

    def __init__(self,name,subject):
        self.subject=subject
        self.name=name
    def teach(self,course):
        self.course=course
        return f"{self.name} is teaching {self.course}"
    def give_assignment(self,ass):
        self.ass=ass
        return f"{self.name} assigned: {self.ass}"
    
class Student:
    def __init__(self,name):
        self.name=name
    def enroll(self,course):
        self.course=course
        return f"{self.name} enrolled in {self.course}"
    def submit_assignment(self,ass):
        self.ass=ass
        return f"{self.name} submitted: {self.ass}"
    
prof = Professor("Dr. Smith", "Computer Science")
stud = Student("Alice")
print(prof.teach("Python"))
print(stud.enroll("Python"))
print(prof.give_assignment("Project 1")) 
print(stud.submit_assignment("Project 1"))
        
```

```python
#  4. Movie Ticket Booking
 
# Create a Cinema class with:
 
# *Attributes: movies (dict → movie: seats available)
 
# *Methods:
 
# * book(movie, seats) → reduces seat count
 
# * cancel(movie, seats) → adds back seats
 
# * show_movies() → shows available seats
 
# Sample Input:
 
# python
 
# cinema = Cinema({"Avatar": 10, "Batman": 5})
 
# print(cinema.book("Avatar", 2))
 
# print(cinema.cancel("Avatar", 1))
 
# cinema.show_movies()
 
# Sample Output:
 
# Booked 2 tickets for Avatar
 
# Cancelled 1 ticket for Avatar
 
# Movies: {'Avatar': 9, 'Batman': 5}
 

class Cinema:
    def __init__(self,movies):
        self.movies=movies
    def book(self,movie,ticket):
        for mov,tic in self.movies.items():
            if mov==movie:
                tic-=ticket
            return f"Booked {ticket} tickets for {movie} \n"
        return f"Tickets not avaliable for {movie} \n"
    def cancel(self,movie,ticket):
        for mov,tic in self.movies.items():
            if mov==movie:
                tic+=ticket
                return f"Cancelled {ticket} tickets for {movie}\n "
        self.movies[movie]=ticket
        return f"Cancelled {ticket} tickets for {movie} \n "
    def show_movies(self):
        print("Movies : ",self.movies)     

cinema = Cinema({"Avatar": 10, "Batman": 5})
 
print(cinema.book("Avatar", 2))
 
print(cinema.cancel("Avatar", 1))
 
cinema.show_movies()

```

```python
# ### 5. Restaurant Menu
 
# Create a Menu class with:
 
# *Attributes: items (dict → item: price)
 
# *Methods:
 
# * add_item(name, price)
 
# * remove_item(name)
 
# * update_price(name, new_price)
 
# * show_menu()
 
# Sample Input:
 
# python
 
# menu = Menu()
 
# print(menu.add_item("Burger", 100))
 
# print(menu.add_item("Pizza", 200))
 
# print(menu.update_price("Pizza", 250))
 
# print(menu.remove_item("Burger"))
 
# menu.show_menu()

# Sample Output:
# Burger added
 
# Pizza added
 
# Pizza price updated to 250
 
# Burger removed
 
# Menu: {'Pizza': 250}
       
class Menu:
    def __init__(self):
        self.items={}
    def add_item(self,name,price):
        self.items[name]=price
        return f"{name} added"
    def update_price(self,name,price):
        self.items.update([(name,price)])
        return f"{name} price is updated to {price}"
    def remove_item(self,name):
        self.items.pop(name)
        return f"{name} removed"
    def show_menu(self):
        print(f"Menu : {self.items}")
menu = Menu()
 
print(menu.add_item("Burger", 100))
 
print(menu.add_item("Pizza", 200))
 
print(menu.update_price("Pizza", 250))
 
print(menu.remove_item("Burger"))

menu.show_menu()
```
